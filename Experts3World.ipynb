```python
bins = list(df['Experts'].quantile([0, 0.80, 0.85, 0.95, 0.96, 0.97, 0.98, 0.99, 1]))
print (bins)


folium.Choropleth(
    geo_data=countries,
    name='choropleth',
    data=df,
    columns=['Country', 'Experts'],
    key_on='feature.properties.admin',
    fill_color='Accent',
    fill_opacity=0.75,
    line_opacity=0.65,
    legend_name='Yugoslav Experts Abroad (1970)', 
    bins=bins,
    reset=True
).add_to(m)

folium.LayerControl().add_to(m)

m
```
